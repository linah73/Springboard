```python
import sqlite3
from sqlite3 import Error

 
def create_connection(db_file):
    """ create a database connection to the SQLite database
        specified by the db_file
    :param db_file: database file
    :return: Connection object or None
    """
    conn = None
    try:
        conn = sqlite3.connect(db_file)
        print(sqlite3.version)
    except Error as e:
        print(e)
 
    return conn

 
def select_all_tasks(conn):
    """
    Query all rows in the tasks table
    :param conn: the Connection object
    :return:
    """
    cur = conn.cursor()
    
    query1 = """
        SELECT *
        FROM FACILITIES
        """
    cur.execute(query1)
 
    rows = cur.fetchall()
 
    for row in rows:
        print(row)
        
    print("\n")
    print("\n")
    
    """Q10: Produce a list of facilities with a total revenue less than 1000. The output 
    of facility name and total revenue, sorted by revenue. Remember that there's a 
    different cost for guests and members!"""

    print("""Q10: Produce a list of facilities with a total revenue less than 1000. The output 
    of facility name and total revenue, sorted by revenue. Remember that there's a 
    different cost for guests and members!""")
    print("\n")

    query10 = """SELECT f.name as Facility,
        SUM(CASE WHEN memid = 0 THEN f.guestcost * b.slots
             ELSE f.membercost * b.slots END) as Revenue
    FROM `Bookings` as b
        INNER JOIN `Facilities` as f USING(facid)
        INNER JOIN `Members` as m USING(memid)
    GROUP BY f.name
    HAVING Revenue < 1000
    ORDER BY Revenue;"""

    cur.execute(query10)
 
    rows = cur.fetchall()
 
    for row in rows:
        print(row)
  
    """Q11: Produce a report of members and who recommended them in alphabetic surname,firstname order"""
    print("\n")
    print("\n")
    print("Q11: Produce a report of members and who recommended them in alphabetic surname,firstname order")
    print("\n")

    query11 = """SELECT m.firstname, m.surname, x.firstname ||' '|| x.surname AS recommendedbyName
    FROM Members m
    LEFT JOIN Members x ON m.recommendedby = x.memid
    WHERE x.memid != 0"""

    cur.execute(query11)
 
    rows = cur.fetchall()
 
    for row in rows:
        print(row)

    """Q12: Find the facilities with their usage by member, but not guests"""
    print("\n")
    print("\n")
    print("Q12: Find the facilities with their usage by member, but not guests")
    print("\n")

    query12 = """SELECT f.name as Facility,  m.firstname ||' '|| m.surname as Member,
        SUM(b.slots) as Facility_Usage
    FROM `Bookings` as b
        INNER JOIN `Facilities` as f USING(facid)
        INNER JOIN `Members` as m USING(memid)
    GROUP BY f.name, b.memid
    HAVING b.memid != 0;"""

    cur.execute(query12)
 
    rows = cur.fetchall()
 
    for row in rows:
        print(row)

    """Q13: Find the facilities usage by month, but not guests"""
    print("\n")
    print("\n")
    print("Q13: Find the facilities usage by month, but not guests")
    print("\n")

    query13 ="""SELECT f.name as Facility, strftime('%m', b.starttime) as Month,
        SUM(b.slots) as Facility_Usage
    FROM `Bookings` as b
        INNER JOIN `Facilities` as f USING(facid)
        INNER JOIN `Members` as m USING(memid)
    WHERE b.memid != 0
    GROUP BY f.name, Month;"""
      
    cur.execute(query13)
 
    rows = cur.fetchall()
 
    for row in rows:
        print(row)

def main():
    database = "sqlite_db_pythonsqlite.db"
 
    # create a database connection
    conn = create_connection(database)
    with conn: 
        print("2. Query all tasks")
        select_all_tasks(conn)
 
 
if __name__ == '__main__':
    main()

```
